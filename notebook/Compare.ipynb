<h3>3. 潜在空間の比較</h3>

<p>&emsp;2つの潜在空間を比較すると、オートエンコーダーは原点から各点群が放射線状に広がるようにして分布しており、VAEは原点を中心としてある程度まとまった分布であった。また、オートエンコーダはVAEに比べ、広い範囲に分布が広がっていることが確認できた。</p>

<p>&emsp;上述した違いは、オートエンコーダーの潜在空間は入力データの特徴量を表現するようになっているが、VAEの潜在空間は潜在空間を入力データに対する確率分布として考えていることによって発生していると考えられる。<br>&emsp;オートエンコーダは、活性化関数にReLU関数を用いているため、出力値の最小値は0となる。このため、原点を中心として放射線状に広がるようにして分布していると考えられる。一方で、VAEはエンコーダによって平均値と分散を求め、それによって定義される確率分布に従うようにして分布する。このため、VAEでは各点群がオートエンコーダーよりも狭い範囲で分布していると考えられる。</p>
